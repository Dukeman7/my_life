```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar datos
df = pd.read_csv('expediente_mensual_maxicable_final.csv')

# 2. Configurar el grfico
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 11))
plt.style.use('seaborn-v0_8-whitegrid')

# Grfico 1: Ingresos vs Tasa BCV
ax1.plot(df['Etiqueta'], df['Ingresos_Brutos_Bs_Est'], color='#1f77b4', marker='o', label='Ingresos (Bs)')
ax1_twin = ax1.twinx()
ax1_twin.plot(df['Etiqueta'], df['Tasa_BCV_Est'], color='#d62728', linestyle='--', marker='x', label='Tasa BCV')
ax1.set_title('MAXI CABLE: Ingresos vs Devaluacin (2024-2025)', fontsize=15, fontweight='bold')
ax1.set_ylabel('Bolvares (Bs)')
ax1_twin.set_ylabel('Tasa BCV (Bs/USD)')
ax1.tick_params(axis='x', rotation=45)

# Grfico 2: Clientes Estimados
ax2.bar(df['Etiqueta'], df['Clientes_Estimados'], color='#2ca02c', alpha=0.7)
ax2.set_title('Estimacin de Clientes Mensuales (ARPU $8)', fontsize=15, fontweight='bold')
ax2.set_ylabel('Nro. de Clientes')
ax2.set_ylim(0, df['Clientes_Estimados'].max() * 1.3)
ax2.tick_params(axis='x', rotation=45)

# Lnea de Promedio
avg_val = df['Clientes_Estimados'].mean()
ax2.axhline(avg_val, color='black', linestyle=':', label=f'Promedio: {int(avg_val)}')

plt.tight_layout()
plt.show()
```
